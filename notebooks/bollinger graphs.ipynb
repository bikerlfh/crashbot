```python
# Definir una lista de precios
precios = [10, 12, 15, 14, 16, 18, 20, 22, 24, 22, 20, 18, 16]

# Convertir la lista de precios en una matriz NumPy de tipo float64
precios_np = np.array(precios, dtype='float64')

# Calcular la media móvil de 20 periodos
sma = talib.SMA(precios_np, timeperiod=20)

# Calcular el oscilador estocástico
k, d = talib.STOCH([max(precios[i:i+1]) for i in range(len(precios)-13)],
                   [min(precios[i:i+14]) for i in range(len(precios)-13)],
                   precios[13:])

# Crear una figura de Matplotlib
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Graficar los precios y la media móvil
axs[0].plot(precios, label='Precios')
axs[0].plot(sma, label='Media móvil de 20 periodos')
axs[0].legend()

# Graficar el oscilador estocástico
x = np.arange(13, len(precios))
axs[1].plot(x, k, label='%K')
axs[1].plot(x, d, label='%D')
axs[1].legend()

# Agregar niveles de sobrecompra y sobreventa
axs[1].axhline(y=80, linestyle='--', color='red', label='Nivel de sobrecompra')
axs[1].axhline(y=20, linestyle='--', color='green', label='Nivel de sobreventa')
axs[1].legend()

# Mostrar la figura
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que los precios están en una lista llamada precios
precios = [0, -1, 0, -1, 0, 1, 2, 1,2,1]

# Convierte la lista de precios en una serie de pandas
serie_precios = pd.Series(precios)

# Calcula la media de los precios
media_precios = serie_precios.mean()
print(f"La media de los precios es: {media_precios}")

# Crea un gráfico de línea de los precios
plt.plot(serie_precios)
plt.axhline(y=media_precios, color='r', linestyle='--', label="Media")
plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.title('Evolución de precios')
plt.legend()
plt.show()
```

```python
# Calcula la media móvil simple de 5 períodos
media_movil_simple = serie_precios.rolling(window=5).mean()

# Añade la media móvil simple al gráfico
plt.plot(serie_precios, label="Precios")
plt.plot(media_movil_simple, label="Media móvil simple (5 períodos)", linestyle="--")
plt.axhline(y=media_precios, color='r', linestyle='--', label="Media")
plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.title('Evolución de precios con media móvil simple')
plt.legend()
plt.show()
```

```python
def detectar_resistencia_soporte(precios, ventana=5, factor_umbral=0.1):
    resistencias = []
    soportes = []

    for i in range(ventana, len(precios) - ventana):
        max_local = max(precios[i - ventana:i + ventana])
        min_local = min(precios[i - ventana:i + ventana])

        umbral = (max_local - min_local) * factor_umbral
        if precios[i] == max_local and max_local - precios[i - ventana] > umbral:
            resistencias.append((i, precios[i]))
        elif precios[i] == min_local and precios[i - ventana] - min_local > umbral:
            soportes.append((i, precios[i]))

    return resistencias, soportes

resistencias, soportes = detectar_resistencia_soporte(precios)

# Graficar precios, media móvil simple, resistencias y soportes
plt.plot(serie_precios, label="Precios")
plt.plot(media_movil_simple, label="Media móvil simple (5 períodos)", linestyle="--")
#plt.axhline(y=media_precios, color='r', linestyle='--', label="Media")
print(resistencias)
for r in resistencias:
    plt.axhline(y=r[1], color='g', linestyle='--', linewidth=0.8)
    plt.scatter(r[0], r[1], color='g', label="Resistencia" if r == resistencias[0] else None)

for s in soportes:
    plt.axhline(y=s[1], color='orange', linestyle='--', linewidth=0.8)
    plt.scatter(s[0], s[1], color='orange', label="Soporte" if s == soportes[0] else None)

plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.title('Evolución de precios con resistencias y soportes')
#plt.legend()
plt.show()
```

```python
def find_support_resistance(prices, window=3):
    support = []
    resistance = []

    for index in range(window, len(prices) - window):
        price_range = prices[index - window : index + window + 1]

        # Identificar máximos y mínimos locales
        max_price = max(price_range)
        min_price = min(price_range)

        if prices[index] == max_price:
            resistance.append((index, max_price))
        elif prices[index] == min_price:
            support.append((index, min_price))

    return support, resistance

# Encontrar niveles de soporte y resistencia en los precios
support, resistance = find_support_resistance(precios_np)

# Graficar los precios, la media móvil simple y las Bandas de Bollinger
plt.plot(serie_precios, label="Precios")
plt.plot(media_movil_talib, label="Media móvil simple (5 períodos) - TA-Lib", linestyle="--")
plt.plot(upper, label="Banda superior - Bollinger", linestyle="--", color='g')
plt.plot(middle, label="Banda media - Bollinger", linestyle="--", color='orange')
plt.plot(lower, label="Banda inferior - Bollinger", linestyle="--", color='r')

# Graficar las líneas de soporte y resistencia
print(support)
for s in support:
    plt.axhline(y=s[1], color='m', linestyle='--', linewidth=0.8, label="Soporte" if s == support[0] else None)

    
print(resistance)
for r in resistance:
    plt.axhline(y=r[1], color='c', linestyle='--', linewidth=0.8, label="Resistencia" if r == resistance[0] else None)

plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.title('Evolución de precios con soporte y resistencia')
plt.legend()
plt.show()
```

```python
import talib

# Definir una lista de precios
precios = [10, 12, 15, 14, 16, 18, 20, 22, 24, 22, 20, 18, 16]
np_precios = np.array(precios, dtype='float64')
# Calcular la media móvil de 20 periodos
sma = talib.SMA(np_precios, timeperiod=20)

# Calcular el oscilador estocástico
k, d = talib.STOCH([max(np_precios[i:i+14]) for i in range(len(np_precios)-13)],
                   [min(np_precios[i:i+14]) for i in range(len(np_precios)-13)],
                   precios[13:])

# Determinar si la tendencia es alcista o bajista
ultimo_precio = precios[-1]
media_movil = sma[-1]
ultimo_k = k[-1]
ultimo_d = d[-1]

if ultimo_precio > media_movil and ultimo_k > ultimo_d:
    print('La tendencia es alcista.')
elif ultimo_precio < media_movil and ultimo_k < ultimo_d:
    print('La tendencia es bajista.')
else:
    print('La tendencia es neutra.')
```
