```python
import pyautogui

# mover el mouse a la coordenada (x=100, y=100) en la pantalla
pyautogui.moveTo(100, 100)

# mover el mouse a la coordenada (x=200, y=200) en 2 segundos
pyautogui.moveTo(1000, 1000, duration=1)

# mover el mouse 100 píxeles hacia la derecha y 100 píxeles hacia abajo
pyautogui.moveRel(100, 100)

# mover el mouse 50 píxeles hacia arriba y 50 píxeles hacia la izquierda en 1 segundo
pyautogui.moveRel(-50, -50, duration=1)
screenWidth, screenHeight = pyautogui.size()
```
